```python
import networkx as nx
import matplotlib.pyplot as plt

# set the number of nodes
num_nodes = 100

# generate random network（Erdős-Rényi）
random_graph = nx.erdos_renyi_graph(num_nodes, p=0.1)

# small world network（Watts-Strogatz）
small_world_graph = nx.watts_strogatz_graph(num_nodes, k=4, p=0.1)

# set layout
layout = nx.spring_layout

# draw random network
plt.figure(figsize=(20, 12))

plt.subplot(121)
nx.draw(random_graph, pos=layout(random_graph), with_labels=True, node_size=500, node_color='skyblue', edge_color='gray')
plt.title("Random Network (Erdős-Rényi)")

# draw small world network
plt.subplot(122)
nx.draw(small_world_graph, pos=layout(small_world_graph), with_labels=True, node_size=500, node_color='lightgreen', edge_color='gray')
plt.title("Small World Network (Watts-Strogatz)")

# show the plot
plt.show()
```
